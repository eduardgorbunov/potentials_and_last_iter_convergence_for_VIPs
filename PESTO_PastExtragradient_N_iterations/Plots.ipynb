```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.io
```

# PEG and OG

```python
res = scipy.io.loadmat('unconstrained.mat')
list_n=np.append(np.array([]),res['N_list'][0])
wc_og3=np.append(np.array([]),res['OG_wc3'][0])
wc_peg3=np.append(np.array([]),res['PEG_wc3'][0])
wc_og4=np.append(np.array([]),res['OG_wc4'][0])
wc_peg4=np.append(np.array([]),res['PEG_wc4'][0])

fig = plt.figure(figsize=(12,8))

plt.semilogy(list_n, wc_og3, linewidth=2, linestyle='dashed',
             markersize = 15, marker = "o", color="forestgreen",
             label=r'OG (step 1/3/L)')
plt.semilogy(list_n, wc_og4, linewidth=2,
             markersize = 15, marker = "o", color="forestgreen",
             label=r'OG (step 1/4/L)')
plt.semilogy(list_n, wc_peg3, linewidth=2,linestyle='dashed',
             markersize = 15, marker = "o", color="blue",
             label=r'PEG (step 1/3/L)')
plt.semilogy(list_n, wc_peg4, linewidth=2,
             markersize = 15, marker = "o", color="blue",
             label=r'PEG (step 1/4/L)')

plt.semilogy(list_n, 5/list_n, linewidth=2,
             markersize = 15, marker = "o", color="orange",
             label=r'$5/k$')

plt.legend(loc='best', fontsize=25)
plt.xlabel(r"$k$", fontsize=35)

plt.xticks(fontsize=30)
_ = plt.yticks(fontsize=30)

plt.grid()
plt.loglog()

ax = fig.gca()
ax.xaxis.offsetText.set_fontsize(28)
ax.yaxis.offsetText.set_fontsize(28)

plt.savefig("PEG_OG.pdf", bbox_inches='tight')
```

# Projected PEG and Projected OG

```python
res = scipy.io.loadmat('constrained.mat')
list_n=np.append(np.array([]),res['N_list'][0])
wc_og=np.append(np.array([]),res['OG_wc4'][0])
wc_peg=np.append(np.array([]),res['PEG_wc4'][0])

fig = plt.figure(figsize=(12,8))

plt.semilogy(list_n, wc_og, linewidth=2,
             markersize = 15, marker = "o", color="forestgreen",
             label=r'Proj OG (step 1/4/L)')
plt.semilogy(list_n, wc_peg, linewidth=2,
             markersize = 15, marker = "o", color="blue",
             label=r'Proj PEG (step 1/4/L)')

plt.semilogy(list_n, 1/list_n, linewidth=2,
             markersize = 15, marker = "o", color="orange",
             label=r'$1/k$')

plt.legend(loc='best', fontsize=25)
plt.xlabel(r"$k$", fontsize=35)

plt.xticks(fontsize=30)
_ = plt.yticks(fontsize=30)

plt.grid()
plt.loglog()

ax = fig.gca()
ax.xaxis.offsetText.set_fontsize(28)
ax.yaxis.offsetText.set_fontsize(28)

plt.savefig("projected_PEG_OG.pdf", bbox_inches='tight')
```
